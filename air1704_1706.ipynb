# FILL NULL

```python
#check null values in pm2.5 
aq_null=aq[aq['PM2.5'].isnull()]
#aq_null
```

```python

unistation=aq.stationId.unique().tolist()
#unistation
```

```python
stations=[]
for i in unistation:
    stations.append(aq.loc[aq['stationId'] == i])
len(stations)
stations
```

```python
#fill pm2.5 and o3
for i in range(len(stations)):
    
#    # stations[i]['PM2.5'] =stations[i]['PM2.5'].fillna(stations[i]['PM2.5'].mean())
#    # stations[i]['O3'] =stations[i]['O3'].fillna(stations[i]['O3'].mean())
    stations[i]['PM2.5']=stations[i]['PM2.5'].fillna(method='ffill').fillna(method='bfill')
    stations[i]['O3']=stations[i]['O3'].fillna(method='ffill').fillna(method='bfill')
    stations[i]['SO2']=stations[i]['SO2'].fillna(method='ffill').fillna(method='bfill')
    stations[i]['NO2']=stations[i]['NO2'].fillna(method='ffill').fillna(method='bfill')
    stations[i]['CO']=stations[i]['CO'].fillna(method='ffill').fillna(method='bfill')
```

```python
#check null for pm2.5 and o3
# for i in range(len(stations)):
#     print(stations[i].isnull().sum())
```

```python
#combining 35 mini df together
for i in range(len(unistation)):
    if i==0:
            prev=stations[i]
            continue
    else:
        result =prev.append(stations[i])
        prev=result
result.head(50)
result.info()
```

```python
total_index=result.index.tolist()
```

```python
result.isnull().sum()
```

```python
#find index of all the null in pm10
pm10_null_index=result[result['PM10'].isnull()].index.tolist()
pm10_null_index
```

```python
#non null pm10 index
pm10_nonnull_index=set(total_index)-set(pm10_null_index)

len(pm10_nonnull_index)
```

```python
#find nonnull pm10
pm10_nonnull_list=[]
for i in pm10_nonnull_index:    
    pm10_nonnull_list.append(result['PM10'][i])
len(pm10_nonnull_list)
#pm10_nonnull_list
```

```python
# #find corresponding pm2.5 to nonnull pm10
pm25_nonnull_list=[]
for i in pm10_nonnull_index:    
    pm25_nonnull_list.append(result['PM2.5'][i])
len(pm25_nonnull_list)
pm25_nonnull_list
```

```python
###############find corresponding pm2.5 to nonnull pm10
# pm25_nonnull_list=[]
# for i in pm10_nonnull_index:    
#     pm25_nonnull_list.extend([[result['PM2.5'][i],result['CO'][i]]])
# len(pm25_nonnull_list)
# #pm25_nonnull_list
```

```python
#find corresponding pm2.5 to null pm10
pm25_list=[]
for i in pm10_null_index:    
    pm25_list.append(result['PM2.5'][i])
len(pm25_list)
```

```python

pm25_nonnull_array=np.array(pm25_nonnull_list)
pm10_nonnull_array=np.array(pm10_nonnull_list)
pm25_array=np.array(pm25_list)
X_train=pm25_nonnull_array
y_train=pm10_nonnull_array
X_test=pm25_array
X_train= X_train.reshape(-1, 1)
y_train= y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)


reg = LinearRegression()
reg.fit(X_train,y_train)

y=reg.predict(X_test)
y_list=y.tolist()
len(y_list)
```

```python
X_train.shape
```

```python
from sklearn.model_selection import KFold, cross_val_score, train_test_split
n_folds = 3
def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(X_train)
    rmse= -cross_val_score(model, X_train,y_train, scoring="neg_mean_absolute_error", cv = kf)
    return(rmse)
```

```python
rmsle_cv(reg)
```
